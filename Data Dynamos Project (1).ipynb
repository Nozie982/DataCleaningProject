```python
import pandas as pd
import matplotlib.pyplot as plt
 
# Load the CSV filefile_path = '/path/to/your/StudentPerformanceFactors.csv'
df = pd.read_csv("C:\\Users\\Capaciti\\Downloads\\StudentPerformanceFactors.csv")
# Count the frequency of each level of parental involvement
parental_involvement_counts = df['Parental_Involvement'].value_counts()
 
# Create a pie chart for parental involvement
plt.figure(figsize=(8, 6))
plt.pie(parental_involvement_counts,labels=parental_involvement_counts.index,autopct='%1.1f%%',startangle=140,colors=plt.cm.Paired.colors)
plt.title("Parental Involvement Levels")
plt.legend(title = "Parental involvement levels:")
plt.show()
# Save the combined figure
# Save the pie chart as an image file
plt.savefig("C:\\Users\\Capaciti\\Downloads\\pie_chart.png")

```
